```python
#Import Mongo Client and parser for the output
from pymongo import MongoClient
from bson.json_util import dumps
```

```python
#Define connection parameters
username = 'omar'
password = 'usfomar'
database = 'training'

url = f'mongodb://{username}:{password}@localhost:27017'
#create  the connection
connection = MongoClient(url)
#list all database names
connection.list_database_names()
training = connection.training
```

```python
training.list_collection_names()
```

```python
#Create a collection called python

collection_name='python'

#drop the  collection if exists
if collection_name in training.list_collection_names():
    training.get_collection(collection_name).drop()

training.create_collection(collection_name)
collection = training.python
```

```python
doc = {"lab":"Accessing mongodb using python", "Subject":"No SQL Databases"}
print(f"Inserting a document into a collection")
try:
    collection.insert_one(doc)
except:
    print(f"Insertion Failed")
else:
    print(f"Inserting completed successfully")

cursor = collection.find()
print(dumps(cursor,indent = 4))

connection.close()

```

### Some Practices on pandas and we can say simple web scraping
1. Extract table that contains the highes worldwide gross movies in the history from url = `https://en.wikipedia.org/wiki/List_of_highest-grossing_films`
2. Do Some Cleaning to the data frame and make it more readable for human beings

#### Import Libraries that i will use

```python
import pandas as pd
import numpy as np
```

**Extract table from the url**

```python

url = 'https://en.wikipedia.org/wiki/List_of_highest-grossing_films'
tables = pd.read_html(url)
tables[0].head()
```

**Copy the the data in a data frame called df for more easier backups (if needed)**

```python
df = tables[0].copy()
df.head()
```

**Remove unneeded columns while make the indicies of the data frame to be the rank of the movies it helps**

```python
columns = ['Peak','Ref']
df.set_index('Rank',inplace = True)#Set the column Rank to be the index of the movies
df.drop(columns=columns, axis=1,inplace=True)#Removes columns i don't need such as Peak and Ref
df.head()
```

```python
#TO KNOW THE DATA TYPES OF ATTRIBUTES OF THE DATA FRAME
df.dtypes
```

**Return the unique values to discover if something odd exists or not that helps me for cleaing process**

```python
df['Worldwide gross'].unique()
```

**I learned RegEx and i want to apply it to improve my understanding of it and it was so easy to just extract the numbers from the gross column which must containt only numbers that represent how much does a movie return  (USD)**

```python
import re
pattern = r'\d'#Using RegEx to parse or extract only the numbers from the gross column to clean it
def parse(x):
    y = re.findall(pattern,x)
    x = ''.join(y)
            
    return x
```

**Using apply function to apply the function i wrote above to the column Worldwide gross**

```python
#Apply the function that extract only numbers from worldwide gross column
df['Worldwide gross'] = df['Worldwide gross'].apply(parse)
```

**Change the data type of that column to be integer and  convert the values to be in billions USD and rounded by 2 decimal points for more readability**

```python
df['Worldwide gross']  = df['Worldwide gross'].astype(int)
df['Worldwide gross'] = np.round(df['Worldwide gross']/1000000000,2)#Convert the values to be in billions USD
```

```python
#Change the column name to Be Worldwide Gross (Billion USD)
df.rename(columns={'Worldwide gross':'Total Gross (Billion USD)'},inplace=True)
df.head()
```

```python
#To display movies in ascending order
df.sort_values('Total Gross (Billion USD)')
```

```python
#if you want to sort movies by year
df.sort_values('Year')
```

**This line of code groups movies by year and compute the sum and the mean of Gross of movies made each year in the data fram and order them descending by sum values**

- We know from that that year 2017 has the highest year of movies gross in the history

```python
df.groupby('Year')['Total Gross (Billion USD)'].agg([('Sum','sum'),  ('Mean', lambda x: np.round(x.mean(),2))]).sort_values('Sum',ascending=False)
```
